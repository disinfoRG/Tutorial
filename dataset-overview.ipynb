```python
import zeroloader as zl
import json
```

```python
ltn = zl.load_data("5030a77a-81fe-11ea-8627-f23c92e71bad", "2020-11", 'service.json')
```

```python
ltn.columns
```

```python
one = ltn[ltn['id']=='1d129c8e-a21c-49a5-9d71-7bde39c38c18'].copy()
```

## Metdata
    - canonical_url: 文章的url
    - metadata: HTML中所包含的metadata，例如：ga-id, opengraph 等

```python
one[["canonical_url", "metadata"]]
```

```python
one['canonical_url'].iloc[0]
```

```python
print(json.dumps(one['metadata'].iloc[0], indent=2, ensure_ascii=False))
```

## Time related fields
    - published_at: 文章發佈時間
    - first_seen_at: 0achive 爬蟲首次抓到該文章的時間
    - last_updated_at: 0archive 爬蟲最後抓到該文章的時間
    - version: 這一筆資料抓到的時間，利用 unix timestamp


```python
one[['published_at', 'first_seen_at', 'last_updated_at', 'version']]
```

```python
from datetime import datetime

one['version_datettime'] = one['version'].apply(lambda t: datetime.fromtimestamp(t))
one[['published_at', 'first_seen_at', 'last_updated_at', 'version', 'version_datettime']]
```

## Content
    - author: 文章作者
    - title: 文章標題
    - text: 內文。公開資料集只提供前280字
    - urls: 文章包含的 url
    - image_urls: 文章包含的圖片 url

```python
one[['author', 'title', 'text', 'urls', 'image_urls']]
```
